```python
'''Problem Statement 1:Unique Event Participants
 A college has a list of students who registered multiple times by mistake.
 Write program using set store participant names and display only unique
  participants'''
# Sample list of participant names (with duplicates)
participants = ["Ayush", "Bedang", "Ayush", "Raghav", "Bedang", "Arjit"]
# Use a set to store unique names
unique_participants = set(participants)
# Display the unique participants
print("Unique participants:")
for name in sorted(unique_participants):
    print(name)
```

```python

'''problem 3:one set contains students in the cricket club,
another in football club
find students who are:
1.in both clubs
2.in only one club
'''
# Define sets for each club
cricket_club = {"Ayush", "Bedang", "kavi","Messi"}
football_club = {"Bedang", "kavi", "Disha","Dhoni"}
both_clubs = cricket_club & football_club
only_one_club = cricket_club ^ football_club
print("Students in both clubs:", both_clubs)
print("Students in only one club:", only_one_club)
```

```python
#adding & updating values in  a dictionary
thisdict={
    "brand":"ferrari",
    "model":"G7",
    "year":1884,
}
thisdict["color"]="RED"
thisdict["model"]="G8"
print(thisdict)
```

```python
'''1) A dictionary stores product names and their prices. Write a program to
update the price of a given product'''
# Dictionary storing product names and their prices
products = {
    "apple": 1.0,
    "banana": 0.5,
    "orange": 0.8
}

# Get user input for the product to update and the new price
product_to_update = input("Enter the product name to update: ")
new_price = float(input("Enter the new price: "))

# if the product exists & update the price
if product_to_update in products:
    products[product_to_update] = new_price
    print(f"Price of {product_to_update} updated to {new_price}")
else:
    print(f"Product '{product_to_update}' not found in the dictionary.")

# Printing the updated dictionary
print("Updated products dictionary:")
print(products)
```

```python
''' 2)A dictionary stores student names  and marks.
write a python program to search for a studnet's marks using name and if name
not found display a suitabke message accrodingly'''
students = {
    "Arnav": 85,
    "Bedang": 92,
    "Sukrut": 78,
    "Dhruv": 88
}
# Get user input:-
student_name = input("Enter the student name to search: ")
# Check if the student exists and display marks/message:-
if student_name in students:
    print(f"{student_name}'s marks: {students[student_name]}")
else:
    print(f"Student '{student_name}' not found in the records.")
# Print the students dictionary:-
print("Students dictionary:")
print(students)
```

```python
'''3. a dictionary stores usernames and passwords. Write a python program to
validate user login'''
# Dictionary storing usernames and passwords
users = {
    "anik": "password123",
    "arjit": "pass1",
    "kushal": "pass2"
}
# Get user input for username and password
username = input("Enter username:")
password = input("Enter password:")
# Validate login
if username in users:
    if users[username] == password:
        print("Login successful!")
    else:
        print("Invalid password.")
else:
    print("Username not found.")
# Print the users dictionary for reference
print("Users dictionary:")
print(users)
```

```python
'''4. a dictionary stores usernames and marks. Write a python program to
find studnet who scored the highest marks'''
# Dictionary with student names and marks
students = {
    "Arjit": 85,
    "anik": 92,
    "Arjun": 78,
    "Bedang": 88,
    "Ayush": 95
}
#Find the highest mark:
highest_marks = max(students.values())
#Find who got that mark (without list comprehension)
topper = []
for name in students:
    if students[name] == highest_marks:
        topper.append(name)
#Showing results:
if len(topper) == 1:
    print("Top student:", topper[0], "->", highest_marks, "marks")
else:
    print("Top students:", ",".join(topper), "->", highest_marks, "marks each")
```
